**Objetivo:** identificar os principais motivos dos clientes estarem cancelando o cartão de crédito.





```python
import pandas as pd
```

```python
#Passo 1 - Importar CSV
tabela = pd.read_csv("/Minicurso Análise de Dados no Python/ClientesBanco.csv", encoding='latin1')
#Passo 2 - Tratar os dados - Excluo colunas que podem não estar envolvidas para alcançar o objetivo final
tabela = tabela.drop("CLIENTNUM", axis = 1) #axis = 1 é para excluir a coluna
display(tabela)
```

## Passos

1. Abri a base de dados e tratei excluindo a coluna CLIENTES que não fazia diferença para a análise final
2. Verifiquei se havia alguma parte da base que podia ser retirada para diminuir o tamanho da tabela
3. Limpei as linhas que não faziam diferença com "tabela = tabela.dropna()

## Agora vamos tratar os valores vazios e exibir um resumo das colunas da base de dados


```python
tabela = tabela.dropna() #Retirei uma linha onde o valor era vazio(Categoria Cartão)
display(tabela.info()) #Diz pra gente informações gerais da tabela

display(tabela.describe().round(1))
```

## Vamos analisar como está a divisão entre Clientes x Cancelados

```python
qtd_categoria = tabela["Categoria"].value_counts() #Contar os valores (quantos clientes e quantos cancelados)
display(qtd_categoria)

qtd_categoria_percentual = tabela["Categoria"].value_counts(normalize = True) #Contar os valores (quantos clientes e quantos cancelados)
display(qtd_categoria_percentual)
```

## Temos várias formas de descobrir os motivos de cancelamento
 - Podemos olhar a comparação entre Clientes e Cancelados em cada uma das colunas de nossa base de dados, para ver se essa informação traz algum insight

```python
import plotly.express as px
```

```python
#Para percorrer as linhas da tabela seria:
#for linhas in tabela.index:

for coluna in tabela: #Isto é pra percorrer as colunas
  grafico = px.histogram(tabela, x = coluna, color = "Categoria")
  grafico.show()
```

## Informações retiradas da análise

- Quanto mais produtos contratados um cliente tem, menor a chance dele cancelar
- A maioria possui o cartão BLUE
- Quanto menos usa o cartão, maior a chance de ele cancelar
- Quanto mais transações e maior o valor da transação, menor a chance de ele cancelar
- Quanto maior o número de contatos, maior a chance de ele cancelar (a partir de 2 contatos já ficar de olho)
