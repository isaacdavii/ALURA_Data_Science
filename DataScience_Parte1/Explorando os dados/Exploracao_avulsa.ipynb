```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
notas = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/ml-latest-small/ratings.csv")
notas
```

```python
notas.columns = ["usuarioId", "filmeId", "nota", "momento"]
notas.head()
```

```python
# Limpeza de dados
notas.drop(columns = "momento", inplace = True) # inplace = True faz a alteração no próprio dataframe
notas
```

```python
# Checando valores nulos em cada coluna
notas.isnull().sum()
```

```python
notas.drop(columns = "usuarioId", inplace = True)
notas
```

```python
sns.barplot(x = notas["nota"].value_counts().index, y = notas["nota"].value_counts().values)
plt.pyplot.title("Distribuição das notas")
plt.pyplot.show()
```

```python
tmdb = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/tmdb_5000_movies.csv")
tmdb.head()
```

```python
tmdb.info()
# Verificamos o que pode ser necessário, ou não, para a análise
# Tiraremos "homepage", "overview", "status", "tagline" e "title" por não serem relevantes
# Tiraremos "original_title" por ser uma redundância de "title"
```

```python
tmdb.drop(columns=["homepage", "overview", "status", "tagline", "title", "keywords", "spoken_languages"], inplace=True)
tmdb.columns = ["orçamento", "gênero", "id", "lingua_original", "título", "popularidade", "produtora", "país_produção", "data_lançamento", "receita", "tempo", "votação_média", "nº_votos"]
tmdb.head()
```

```python
tmdb = tmdb.set_index("id")
tmdb.head()
```

```python
tmdb["lingua_original"].value_counts()
```

```python
tmdb.query("receita == 0").shape
# query() é uma função que filtra o dataframe de acordo com a condição passada
# shape retorna o número de linhas e colunas do dataframe
tmdb.shape
```

```python
tmdb = tmdb.query("receita > 0")
# Filtrando os filmes que tiveram receita maior que 0
# Se a receita for 0, o filme não foi lançado
tmdb.shape
```

```python
sns.displot(tmdb["receita"])
plt.title("Distribuição da receita dos filmes")
plt.xlabel("Receita")
plt.ylabel("Quantidade de filmes")
plt.show()
```
